```python
# train_gender_yolo_only.py
# Train YOLOv8 classification model for gender identification

from ultralytics import YOLO
import os

# ----------- CONFIG -----------
DATASET_PATH = "dataset"          # your dataset folder path
TRAIN_DIR = os.path.join(DATASET_PATH, "training")
VAL_DIR = os.path.join(DATASET_PATH, "validation")
EPOCHS = 25                       # adjust if needed
IMG_SIZE = 224
BATCH_SIZE = 32
MODEL_NAME = "gender_yolo_model"  # output folder name
# ------------------------------

print("🚀 Starting YOLOv8 gender classification training...")

# 1️⃣ Load pre-trained YOLOv8 classification model
model = YOLO("yolov8n-cls.pt")  # use yolov8m-cls.pt for better accuracy (more GPU)

# 2️⃣ Train model
model.train(
    data=TRAIN_DIR,      # training folder path
    val=VAL_DIR,         # validation folder path
    epochs=EPOCHS,
    imgsz=IMG_SIZE,
    batch=BATCH_SIZE,
    name=MODEL_NAME
)

# 3️⃣ Validate model after training
print("📊 Evaluating model performance...")
metrics = model.val()
print("📈 Validation Metrics:", metrics)

# 4️⃣ Export trained model (for later testing)
model.export(format="onnx")  # optional: or 'torchscript', 'tflite', etc.
print("✅ Training complete! Model saved at:")
print(f"➡ runs/classify/{MODEL_NAME}/weights/best.pt")
```
